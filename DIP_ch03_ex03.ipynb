# 範例 3-3 直方圖顯示

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
```

讀入影像

```python
url='http://www.cs.albany.edu/~xypan/research/img/Kodak/kodim11.png'
img=io.imread(url)
plt.imshow(img)
plt.axis('off')
```

劃出灰階直方圖

```python
gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.hist(gray.ravel(), 256, [0, 256],color='m')
plt.show()
```

劃出彩色之直方圖

```python
color = ('r','g','b')
for i, col in enumerate(color):
  histr = cv2.calcHist([img],[i],None,[256],[0, 256])
  plt.plot(histr, color = col)
  plt.xlim([0, 256])
plt.show()
```
