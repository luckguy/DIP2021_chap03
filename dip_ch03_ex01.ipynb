## 範例 3-1 : Gamma 校正

```python
import cv2
import numpy as np
import imutils
import matplotlib.pyplot as plt
```

讀入影像

```python

url='https://sipi.usc.edu/database/preview/misc/boat.512.png'
img=imutils.url_to_image(url)
plt.figure(figsize=(14, 14), dpi=100)
fig=plt.imshow(img)
plt.axis('off')
```

利用**指數運算**進行 Gamma 校正

```python

gamma=2
gray_img1=img/255.   # 正規化
corrected=np.power(gray_img1,gamma)

plt.figure(figsize=(16,16),dpi=100)
plt.subplot(121)
plt.imshow(img)
plt.axis('off')
plt.subplot(122)
plt.imshow(corrected)
plt.axis('off')
```
